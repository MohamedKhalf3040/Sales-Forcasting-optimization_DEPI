### Store 1 AUTOMOTIVE Family data

```python
auto_store1 = traing_set[(traing_set['family'] == 'AUTOMOTIVE') & (traing_set['store_nbr'] == 1)]
auto_store1 = auto_store1.drop(['id', 'date', 'store_nbr', 'family', 'year', 'month'], axis=1)
auto_store1
```

### Exp. 1: RNN 

```python
# Function to create sequences
def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data)-(seq_length)):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)
```

```python
# Define sequence length
SEQ_LENGTH = 7  # Using past 5 days to predict the next day

X, y = create_sequences( auto_store1.values, SEQ_LENGTH)
```

```python
# Split into training and testing sets (80-20 split)
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
def build_simple_rnn(input_shape):
    inputs = Input(shape=input_shape)
    x = SimpleRNN(255, activation='tanh')(inputs)
    outputs = Dense(3)(x)
    model = Model(inputs, outputs)
    model.compile(optimizer='adam', loss='mse')
    return model

# Build the model
simple_rnn = build_simple_rnn((7,3))
simple_rnn.summary()
```

```python
# Train the model
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=20
)
history_rnn_auto_store1 = simple_rnn.fit(X_train, y_train, epochs=1000, batch_size=32,
                             validation_split=0.2, callbacks=[early_stop])
```

```python
# summarize history for loss
plt.plot(history_rnn_auto_store1.history['loss'])
plt.plot(history_rnn_auto_store1.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```python
# Function to evaluate models
def evaluate_model(model, X_test, y_test):
    predictions = model.predict(X_test)

    mse = mean_squared_error(y_test, predictions)
    mae = mean_absolute_error(y_test, predictions)
    return mse, mae, predictions
```

```python
mse_rnn, mae_rnn, pred_rnn = evaluate_model(simple_rnn, X_test, y_test)
```

```python
# save experiment
# MLFlow init
mlflow.set_experiment('RNN_Exp')
mlflow.set_tracking_uri('http://127.0.0.1:5000')

best_params = {
    'RNN_units': 255
}
metrics = {
    'MSE': mse_rnn,
    'MAE': mae_rnn,
}

mlflow.log_params(best_params)
mlflow.log_metrics(metrics)
mlflow.keras.log_model(simple_rnn, 'RNN Model')
print('model has been registed')
```

### Exper.2: LSTM

```python
def build_lstm(input_shape):
    inputs = Input(shape=input_shape)
    x = LSTM(50)(inputs)
    outputs = Dense(3)(x)
    model = Model(inputs, outputs)
    model.compile(optimizer='adam', loss='mse')
    return model
```

```python
# Build LSTM model
lstm_model = build_lstm((SEQ_LENGTH,3))
lstm_model.summary()
```

```python
# Train LSTM model
history_lstm = lstm_model.fit(X_train, y_train, epochs=100, batch_size=32,
                              validation_split=0.2, callbacks=[early_stop])
```
